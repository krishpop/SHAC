```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
```

```python
# This is using steps_min=4
%matplotlib notebook
data = np.load("outputs/2023-08-09/16-38-28/logs/truncation_analysis.npz")

ax = plt.gca()

norms = data["contact_forces"]
early = data["early_termination"]
contact = data["contact_truncation"]
horizon = data["horizon_truncation"]
ends = data["episode_ends"]
plt.plot(norms)
mi, ma = np.min(norms), np.max(norms)
ax.fill_between(np.arange(len(norms)), mi, ma, where=early.flatten(), color='g', alpha=0.2, label="early")
ax.fill_between(np.arange(len(norms)), mi, ma, where=contact.flatten(), color='r', alpha=0.2, label="contact")
ax.fill_between(np.arange(len(norms)), mi, ma, where=horizon.flatten(), color='b', alpha=0.2, label="horizon")
ax.fill_between(np.arange(len(norms)), mi, ma, where=ends.flatten(), color='y', alpha=0.2, label="end")

ax.legend()
ax.set_ylabel("contact forces norm")
ax.set_xlabel("timestep")

print("early terms: ", np.sum(early))
print("contact terms: ", np.sum(contact))
print("horizon terms: ", np.sum(horizon))
print("episode ends: ", np.sum(ends))

plt.savefig("hopper_cfs.png")
```

```python
%matplotlib notebook
# Ranges for accumulation
ranges = np.arange(0,norms.shape[0], 32)
ranges = np.append(contact, [0, norms.shape[0]])
ranges = np.sort(ranges)

# Initialize an array to store the accumulated values
accumulated_array = np.zeros_like(norms).squeeze()

# Accumulate the array within the specified ranges
for i in range(len(ranges)-1):
    start = ranges[i]
    end = ranges[i+1]
    accumulated_array[start:end] = np.cumsum(norms[start:end])

plt.plot(accumulated_array)
```
